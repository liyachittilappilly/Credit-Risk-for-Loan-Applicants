```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

d = pd.read_csv("credit_risk.csv")
```

```python
# Checks what unique values is in class column
d['class'].unique()
```

```python
X = d.columns.drop("class")
y = d['class']
```

```python
dencoded = pd.get_dummies(d[X])
#  column might have categories (Male/Female')
#get_dummies converts this to numbers (one-hot encoding)
dencoded.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(dencoded, y,test_size=0.15,random_state=100)

print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
train_accuracy = model.score(X_train,y_train)
print("Train accuracy = ", train_accuracy)

test_accuracy = model.score(X_test,y_test)
print("Test accuracy = ", test_accuracy)
```

```python
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

from sklearn.metrics import confusion_matrix

train_conf_matrix = confusion_matrix(y_train,train_predictions)
pd.DataFrame(train_conf_matrix,columns=model.classes_,index=model.classes_)
```

```python
test_conf_matrix = confusion_matrix(y_test,test_predictions)
pd.DataFrame(test_conf_matrix,columns=model.classes_,index=model.classes_)
```

```python
train_correct_predictions = train_conf_matrix[0][0]+train_conf_matrix[1][1]
#adding true negatives and true positives
train_total_predictions = train_conf_matrix.sum() #to no of predictions
train_accuracy = train_correct_predictions/train_total_predictions #manual accuracy of training data
print(train_accuracy)
```

```python
# Calculating test accuracy from confusion matrix
test_correct_predictions = test_conf_matrix[0][0]+test_conf_matrix[1][1]
total_predictions = test_conf_matrix.sum()
test_accuracy = test_correct_predictions/total_predictions
print(test_accuracy)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,test_predictions))
```
